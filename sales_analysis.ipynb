Question 1: What percentage of total orders were shipped on the same date?

```python
print(sqldf(''' SELECT COUNT(row_id) as total_orders,
                SUM(CASE WHEN Order_Date = Ship_Date then 1 else 0 END) AS shipped_same_date,
                ROUND(cast(sum(case when Order_Date = Ship_Date then 1 else 0 END) as float(1,2)) / 
                cast(count(row_id) as float(1,2)) * cast(100 as float(1,2)), 2) as perCE
                from df2
            '''))
```

Question 2: Name top 3 customers with highest total value of orders.

```python
print(sqldf('''
    SELECT Customer_ID, Customer_Name, ROUND(SUM(Sales),2) as Sum_of_Sales
    FROM df2
    GROUP BY Customer_ID, Customer_Name
    ORDER BY Sum_of_Sales DESC
    LIMIT 3
'''))
```

Question 3: Find the top 5 items with the highest average sales per day.

```python
print(sqldf('''
    SELECT 
        Product_Name, 
        ROUND(AVG(Daily_Sales),2) AS Avg_Sales_Per_Day, DATE(Order_Date) as Date
    FROM (
        SELECT 
            Product_Name,
            Order_Date,
            SUM(Sales) AS Daily_Sales
        FROM df2
        GROUP BY Product_Name, Order_Date
    ) AS sub
    GROUP BY Product_Name
    ORDER BY Avg_Sales_Per_Day DESC
    LIMIT 5
'''))
```

Question 4: Write a query to find the average order value for each customer, and rank the customers by their average order value.

```python
print(sqldf('''select Customer_ID, Customer_Name,
               cus_avg_order_value,
               RANK() OVER(ORDER BY cus_avg_order_value desc) as Customer_Rank
               from 
               (
               select Customer_ID, Customer_Name,
               avg(Sales) as cus_avg_order_value
               from df2
               group by Customer_ID, Customer_Name
               )
            '''))
```

Question 5: Give the name of customers who ordered highest and lowest orders from each city.

```python
from pandasql import sqldf

# helper to run sql against pandas DataFrames in local scope
pysqldf = lambda q: sqldf(q, globals())

query = '''
WITH max_per_city AS (
  SELECT city, MAX(Sales) AS high_order_value
  FROM df2
  GROUP BY city
),
min_per_city AS (
  SELECT city, MIN(Sales) AS low_order_value
  FROM df2
  GROUP BY city
)
SELECT 
  h.customer_name AS high_value_customer,
  mp.high_order_value,
  l.customer_name AS low_value_customer,
  np.low_order_value,
  mp.city
FROM max_per_city mp
JOIN df2 h
  ON h.city = mp.city
  AND h.Sales = mp.high_order_value
JOIN min_per_city np
  ON np.city = mp.city
JOIN df2 l
  ON l.city = np.city
  AND l.Sales = np.low_order_value
;
'''

result_df = pysqldf(query)
# result_df now holds the output as a pandas DataFrame
print(result_df)
```

Question 11: Which segment places the highest number of orders from each state and which segment places the largest individual orders from each state?

```python
print(sqldf('''SELECT State, Segment, num_orders 
               from
               (
               SELECT State,Segment,
               count(Order_ID) as num_orders,
               row_number() over(partition by State order by count(Order_ID) desc) as rank_num
               from df2
               group by State, Segment
               )
               where rank_num=1
            '''))
```

Question 12: Find the maximum number of days for which total sales on each day kept rising.

```python
print(sqldf('''SELECT MAX(Days) AS Maximum_Consecutive_Days
FROM (
  SELECT
    a.Order_Date,
    COUNT(*) AS Days
  FROM
    df2 a
  JOIN
    df2 b ON a.Order_Date <= b.Order_Date
  GROUP BY
    a.Order_Date
  HAVING
    SUM(b.Sales) - SUM(a.Sales) = a.Sales
) AS ConsecutiveDays
'''))
```
